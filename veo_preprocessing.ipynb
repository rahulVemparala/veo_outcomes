```python
# Dataset Imports
from google.colab import files

def upload_files():
    """
    Allows the user to upload files to Colab's temporary storage.

    Returns:
        dict: A dictionary where the keys are the filenames and the values are the file contents.
    """
    return list(files.upload().keys())
```

```python
uploaded = []

while True:
    user_input = input("Say 'yes' to start an Upload or type 'done' to finish: ")

    if user_input.lower() == 'done':
        break

    uploaded.append(upload_files()[0])
```

```python
import openpyxl
import pandas as pd

# Load the Excel workbook
wb = openpyxl.load_workbook(uploaded[0])

# Print sheet names
for sheet in wb.sheetnames:
    print(sheet)
```

```python
occ_pay = pd.read_excel(uploaded[0], sheet_name=wb.sheetnames[0], skiprows=3)
occ_ind = pd.read_excel(uploaded[0], sheet_name=wb.sheetnames[2], skiprows=3)
edu = pd.read_excel(uploaded[0], sheet_name=wb.sheetnames[12], skiprows=3)
age = pd.read_excel(uploaded[0], sheet_name=wb.sheetnames[12], skiprows=3)
```

```python
occ_x_grade = pd.merge(occ_ind,
                       occ_pay,
                       on =['cohort', 'label_cohort', 'cohort_years',
                            'dod_occ_code_level', 'dod_occ_code', 'label_dod_occ_code'],
                       how = 'left')
```

```python
occ_x_grade.shape
```

```python
merg_cols = [
 'Obs_x', 'cohort', 'label_cohort', 'cohort_years', 'dod_occ_code_level',
       'dod_occ_code', 'label_dod_occ_code', 'ind_level', 'industry',
       'label_industry', 'y1_emp_x', 'y5_emp_x', 'y10_emp_x',
       'y1_p25_earnings_x', 'y5_p25_earnings_x', 'y10_p25_earnings_x',
       'y1_p50_earnings_x', 'y5_p50_earnings_x', 'y10_p50_earnings_x', 'Obs_y',
       'paygrade_level', 'paygrade', 'label_paygrade', 'y1_emp_y', 'y5_emp_y',
       'y10_emp_y', 'y1_nonemp', 'y5_nonemp', 'y10_nonemp',
       'y1_p25_earnings_y', 'y5_p25_earnings_y', 'y10_p25_earnings_y',
       'y1_p50_earnings_y', 'y5_p50_earnings_y', 'y10_p50_earnings_y',
]

merg_cols.sort()


occ_x_grade = occ_x_grade.filter(items=merg_cols)
```

```python
occ_x_grade.columns = [col.replace(r'_x', '_ind') for col in occ_x_grade.columns]
```
